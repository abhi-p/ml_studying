### Dimension Reduction
- More effcient storage and computation
- Removes less informative 'noise' features

### Principal Component Reduction (PCA)
- Two Steps i) decorrelation ii) dimension reduction

#### i) Decorrelation
    - Rotates data samples to be aligned with axes
    - Shifts data samples so they have mean 0
    - no information is lost
    PCA is a sklearn component
    - fit() learns transformation from given data
    -transform() applies the learned transformation
    - transform() can also be applied to new data
    - samples = array of two features


```python
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA

wine=load_wine()
model=PCA()
model.fit(wine.data)

transformed=model.transform(wine.data)
print(transformed)
print(model.components_)
```

### Intristic dimension
    - intrinsic dimension = number of festures needed to approximate the dataset
    -EX: We might have a 2D dataset of lat,lng that represent a planes flight path. This however can be approximated in 1D by the displacement
    - Essential idea behind dimension reduction
    - What is the most compact representation of the samples
    - Can be detected by PCA

### Versicolor dataset
    - "versicolor", one of the iris species
    - Only 3 features: sepal length, sepal width, petal width
    - Samples are in 3D but can be approximated in 2D
    - Intrisic dimenison= number of PCA features with significant variance
    - PCA features are organized by highest to lowest variance
    - intristic dimension is the number of PCA features with significant variance

### Word frequency arrays
    - Rows represent documents, columns represent words
    - Entries measure the measure of precense of each word in each document
    - tf-idf (term frequency, inverse document frequency)
    - something along the lines of (# of times the term is in document)/*

```python
#A tf-idf word-frequency array

# Import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

documents=['cats say meow', 'dogs say woof', 'dogs chase cats']

# Create a TfidfVectorizer: tfidf
tfidf = TfidfVectorizer() 

# Apply fit_transform to document: csr_mat
csr_mat = tfidf.fit_transform(documents)

# Print result of toarray() method
print(csr_mat.toarray())

# Get the words: words
words = tfidf.get_feature_names_out()
# Print words
print(words)
```
