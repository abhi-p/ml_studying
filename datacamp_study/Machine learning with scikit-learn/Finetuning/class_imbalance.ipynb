## Class Imbalance
- Classification for predicting fradulent bank transactions:
    - 99% of transactions are legitimate; 1% are fraudulent
- Could build a classfier that predicts NONE of the transactions are fradulent
    - 99% accuracy
    - But terrible at actuall predicting fraud
    - Fails at its original purpose
- Class imbalance: uneven frequency of classes

## Confusion Matrix:
- True Positive
    - Actual: Fraudulent
    - Predicted: Fraudulent
- True Negative
    - Actual: Legitimate
    - Predicted: Legitimate
- False Negative:
    - Actual: Fraudulent
    - Predicted: Legitimate
- False Postive:
    - Actual: Legitmiate
    - Predicted: Fraudulent

- Accuracy:
    - (tp+tn)/(tp+tn+fp+fn)
- Precision:
    - (tp)/(tp+fp)
    - High precision= lower false postive rate
        - Not many legitimate transctions are predicted as fraud
- Recall (sensitivity):
    - (tp)/(tp+fn)
    - High Recall = lower false negative rate
        - Most fradulent transations are detected

- Harmonic Mean:
    - F1 Score: 2* (precision*recall)/(precision+recall)
        - Treats precision and recall equally. 
- [[tn, fp], [fn, tp]]
