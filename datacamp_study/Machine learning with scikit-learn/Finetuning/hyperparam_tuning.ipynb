## Hyper Param Tuning
### Chpsing the correct hyper params:
    - Try lots of different hyoer param values
    - Fit all of them separately
    - See how well they perform
    - Choose the best performing values
        - This is called hyper param tuning
        - It is essential to use cross validation to not overfit the test set
        - We can still split the dataset and perform cv on training set
        - We withhold test set for final eval

### Grid Search
