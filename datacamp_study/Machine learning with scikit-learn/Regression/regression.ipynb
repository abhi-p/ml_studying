```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('..\\data\\diabetes_clean.csv')
df.head()
```

```python
X=df.drop('glucose',axis=1).values
y=df['glucose'].values
print(type(X),type(y))
```

```python
X_bmi=df['bmi'].values

print(X_bmi.shape,y.shape)
X_bmi=X_bmi.reshape(-1,1)
print(X_bmi.shape,y.shape)
```

```python
plt.scatter(X_bmi,y)
plt.xlabel('bmi')
plt.ylabel('glucose')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_bmi,y)
prediction=reg.predict(X_bmi)

plt.scatter(X_bmi,y)
plt.plot(X_bmi,prediction)
plt.xlabel('bmi')
plt.ylabel('glucose')
plt.show()

```

### RMSE

```python
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(y,prediction)
```
