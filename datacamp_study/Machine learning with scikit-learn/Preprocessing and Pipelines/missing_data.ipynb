```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
music_df=pd.read_csv('..\\data\\music_clean.csv')
music_df.isna().sum().sort_values()
```

## How to deal with missing data
### Drop missing data:
-  music_df=music_df.dropna(subset=['genre','popularity','liveness','tempo'])
### Impute Missing data
- Use subject matter expertise to replace missing data with educated guessess
- Can use mean, median
- For categorical often mode is used
- Must split our data to avoid data leakage between test and train


```python
#Imputing missing values
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split

#separate the categorical and numerical columns
X_cat=music_df['genre'].values.reshape(-1,1)
X_num=music_df.drop(['genre','popularity'],axis=1).values

#separate the target
y=music_df['popularity'].values

#split the data
X_train_cat,X_test_cat,y_train,y_test=train_test_split(X_cat,y,test_size=0.2,random_state=12)
X_train_num,X_test_num,y_train,y_test=train_test_split(X_num,y,test_size=0.2,random_state=12)

#impute
imp_cat=SimpleImputer(strategy='most_frequent')
X_train_cat= imp_cat.fit_transform(X_train_cat)
X_test_cat= imp_cat.fit_transform(X_test_cat)

imp_num=SimpleImputer(strategy='mean')
X_train_num= imp_num.fit_transform(X_train_num)
X_test_num= imp_num.fit_transform(X_test_num)

#combine the data
X_train=np.append(X_train_cat,X_train_num,axis=1)
X_test=np.concatenate(X_test_cat,X_test_num,axis=1)
```

```python
#Impute within a pipeline

from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split



music_df=music_df.dropna(subset=['genre','popularity','liveness','tempo'])
music_df['genre']=np.where(music_df['genre']=='Rock',1,0)
X=music_df.drop('genre',axis=1).values
y=music_df['genre'].values

steps=[
    ('imputer',SimpleImputer()),
    ('classifier',LogisticRegression())
]
pipeline=Pipeline(steps)

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
pipeline.fit(X_train,y_train)
pipeline.score(X_test,y_test)
```
